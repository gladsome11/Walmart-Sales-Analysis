```python
df['Hour']= pd.to_datetime(df['Time'], format ='%H:%M:%S').dt.hour
```

```python
hourlysales = df.groupby('Time') .agg({'Total':'sum'})
hourlysales.sort_values(by='Total',ascending=False)
```

```python
def categorizetime (hour):
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 23:
        return 'Evening'
    else:
        return 'Midnight'
    
df['Timeperiod']= df['Hour'].apply(categorizetime)   
df
    
```

```python
periodsales = df.groupby('Timeperiod') .agg({'Total':'sum'})
periodsales.sort_values(by='Total',ascending=False)
```

```python
salesovertime = df.groupby('Date').agg({'Total':'sum'}).reset_index()

plt.figure(figsize=(12,6))
plt.plot(salesovertime['Date'],salesovertime['Total'],marker='o')
plt.title('Sales Performance Over Time')
plt.xlabel('Date')
plt.ylabel('Total')
plt.grid(True)
plt.show()


```

## How do customer ratings correlate with sales performance?

### Do higher rated products generate more sales?

```python
product_performance = df.groupby('Product line').agg({'Total':'sum','Rating':'mean'})
product_performance.sort_values(by='Total',ascending=False)
```

```python
pp= product_performance['Total'].corr(product_performance['Rating'])
print('Correlation between total and ratings: ')
print (pp)
```

```python
plt.figure(figsize=(12,6))
plt.hist(df['Rating'],bins=10,color='purple',edgecolor='black')
plt.title('Distribution of Customer Ratings')
plt.xlabel('Customer Ratings')
plt.ylabel('Frequency')
plt.show()
```

## Which Product line are the most profitable and least profitable?

### Most and least profitable product lines?

```python
profit_product = df.groupby('Product line')['gross income'].sum().reset_index()
plt.figure(figsize=(12,6))
plt.bar(profit_product['Product line'],profit_product['gross income'],color='green')
plt.title('Profit By Product Line')
plt.xlabel('Product Line')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

### Top selling product line by revenue and volume

```python
toprevenue = df.groupby('Product line').agg({'Total':'sum'}).reset_index()
toprevenue.sort_values(by=['Total'],ascending=[False])


plt.figure(figsize=(12,6))
plt.bar(toprevenue['Product line'],toprevenue['Total'],color='coral')
plt.title('Top Selling Products by Revenue')
plt.xlabel('Product line')
plt.ylabel('Total')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
topvolume = df.groupby('Product line').agg({'Quantity':'sum'}).reset_index()
topvolume.sort_values(by=['Quantity'],ascending=[False])
plt.figure(figsize=(12,6))
plt.bar(topvolume['Product line'],topvolume['Quantity'],color='orange')
plt.title('Top Selling Products by Volume')
plt.xlabel('Product line')
plt.ylabel('Quantity Sold')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
    
```
